### 4. Download Public Voicepack from Hugging Face
Downloads the finetuned `nepali_ft.pt` voicepack from the correct public Hugging Face repository: `Firoj112/kokoro-nepali-base-adaptation`.

```python
from huggingface_hub import hf_hub_download
import os

repo_id = "Firoj112/kokoro-nepali-base-adaptation"
print(f"Downloading voicepack from public repository: {repo_id}...")
os.makedirs("kikiri-tts/kokoro/voices", exist_ok=True)

try:
    hf_hub_download(
        repo_id=repo_id,
        filename="voices/nepali_ft.pt",
        local_dir="."
    )
    print("Success! Voicepack downloaded to: kikiri-tts/kokoro/voices/nepali_ft.pt")
except Exception as e:
    print(f"Failed to download voicepack: {e}")
```
