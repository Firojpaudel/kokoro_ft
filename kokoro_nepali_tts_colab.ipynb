### 4. Download Public Checkpoint and Voicepack from Hugging Face
Downloads the finetuned `nepali_ft.pt` voicepack and the `nepali_kokoro_epoch3.pth` model checkpoint from your public Hugging Face model repository.

```python
from huggingface_hub import hf_hub_download
import os

repo_id = "Firoj112/kokoro-nepali-base-adaptation"
print(f"Downloading files from public repository: {repo_id}...")

# 1. Download voicepack
os.makedirs("kikiri-tts/kokoro/voices", exist_ok=True)
try:
    hf_hub_download(
        repo_id=repo_id,
        filename="voices/nepali_ft.pt",
        local_dir="."
    )
    print("Success! Voicepack downloaded to: kikiri-tts/kokoro/voices/nepali_ft.pt")
except Exception as e:
    print(f"Failed to download voicepack: {e}")

# 2. Download model checkpoint
os.makedirs("checkpoints/stage2_voice", exist_ok=True)
try:
    hf_hub_download(
        repo_id=repo_id,
        filename="nepali_kokoro_epoch3.pth",
        local_dir="."
    )
    print("Success! Model checkpoint downloaded to: checkpoints/stage2_voice/nepali_kokoro_epoch3.pth")
except Exception as e:
    print(f"Failed to download model checkpoint: {e}")
```
